## –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö

–ö–∞–∂–¥—ã–π –æ—Ç–∫–ª–∏–∫ –æ–ø–∏—Å–∞–Ω 8 –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏:  
* $\Delta$ùëÖ, $\delta$ùëÖ, ùëíùëöùëé0.001ùêº, ùëíùëöùëé0.01ùêº, ùëíùëöùëé0.1ùêº, ùëíùëöùëé0.001ùê∑, ùëíùëöùëé0.01ùê∑, and ùëíùëöùëé0.1ùê∑. –†–µ–≥–∏—Å—Ç—Ä–∏—Ä—É–µ—Ç—Å—è 16 —Å–µ–Ω—Å–æ—Ä–∞–º–∏, –∏—Ç–æ–≥–æ –∫–∞–∂–¥—ã–π –æ—Ç–∫–ª–∏–∫ –æ–ø–∏—Å–∞–Ω 128 –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏.

–ò—Å–ø–æ–ª—å–∑—É–µ–º—ã–µ –∞–Ω–∞–ª–∏—Ç—ã: 
* 1: Ethanol; 2: Ethylene; 3: Ammonia; 4: Acetaldehyde; 5: Acetone; 6: Toluene

–°—Ç—Ä–æ–∫–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ –∞–Ω–∞–ª–∏—Ç–∞ (1-6), –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏, –∏ 128 –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
* 1;10.000000 1:15596.162100 2:1.868245 3:2.371604 ... 127:-0.902241 128:-2.654529 

```python
gas_data = []

voc = {1: 'Ethanol', 2: 'Ethylene', 3: 'Ammonia', 4: 'Acetaldehyd', 5: 'Acetone', 6: 'Toluene'}

with open(path + 'batch1.dat', 'r') as file_data:
    for i, line in enumerate(file_data.readlines()):
        data = line.strip().split(' ') 
        gas_type, conc = data[0].strip().split(';')
        gas_type = voc[int(gas_type)]

        x, y = zip(*[(int(i.strip().split(':')[0]), float(i.strip().split(':')[1])) for i in data[1:]])
        gas_data.append([gas_type] + [conc] + list(y))
```

```python
features = ('R', '‚ÄñŒîR‚Äñ', 'ema0.001l', 'ema0.01l', 'ema0.1l', 'ema0.001D', 'ema0.01D', 'ema0.1D')
sensors = [str(i) for i in range(16)]

df_columns = ['VOC', 'conc']
for i in sensors:
    for j in features:
        df_columns.append('s' + i + '_' + j)


df = pd.DataFrame(gas_data, columns=df_columns)
df.head()
```

```python
df.tail()
```

## PCA

PCA, –∫–æ–ø–∏—è —Å –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Ñ–∞–π–ª–∞. –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç—Å—è –¥–≤–∞ –≤–∞—Ä–∏–∞–Ω—Ç–∞:
* –≤—Å–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã (feat1)
* —Ç–æ–ª—å–∫–æ –æ—Å–Ω–æ–≤–Ω—ã–µ (feat2)

```python
feat1 = df.values[:,2:]

# select columns with main features
p = list(range(len(df_columns)))
main_comp = [v for i, v in enumerate(p) if (i - 2) % 8 in {0, 1}]
feat2 = df.values[:,main_comp]
```

–ó–Ω–∞—á–µ–Ω–∏—è –≤ —Å–≤–æ–µ–π –º–∞—Å—Å–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç "–æ–∫–æ–ª–æ –Ω—É–ª—è" –¥–æ –∫–∞–∫–æ–≥–æ-—Ç–æ –º–∞–∫—Å–∏–º—É–º–∞, –∏ —É–º–µ—Å—Ç–Ω–µ–µ, –ø–æ—ç—Ç–æ–º—É, –¥–∞–Ω–Ω—ã–µ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞—Ç—å –ø–æ MaxAbsScaler.

(StandardScaler / MinMaxScaler / MaxAbsScaler https://scikit-learn.ru/6-3-preprocessing-data/ )    

```python
from sklearn.decomposition import PCA as sk_pca
# from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler

sc = MaxAbsScaler() 

nfeat1 = sc.fit_transform(feat1)
nfeat2 = sc.fit_transform(feat2)

skpca1 = sk_pca(n_components=2)
skpca2 = sk_pca(n_components=2)
 
# Transform on the scaled features
Xt1 = skpca1.fit_transform(nfeat1) 
Xt2 = skpca2.fit_transform(nfeat2) 
```

```python
# Define the labels for the plot legend
labplot = list(voc.values()) 
 
# Scatter plot
colors = [plt.cm.jet(float(i)/len(labplot)) for i in range(len(labplot))]

with plt.style.context(('ggplot')):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9,6))
    fig.set_tight_layout(True)
 
    for i, u in enumerate(labplot):
        col = np.expand_dims(np.array(colors[i]), axis=0)
        xi = [Xt1[j,0] for j in range(len(Xt1[:,0])) if df['VOC'].iloc[j] == u]
        yi = [Xt1[j,1] for j in range(len(Xt1[:,1])) if df['VOC'].iloc[j] == u]
        ax1.scatter(xi, yi, c=col, s=60, edgecolors='k',label=str(u))

    
    ax1.set_xlabel("PC1")
    ax1.set_ylabel("PC2")
    ax1.set_title('All features')
 
    for i, u in enumerate(labplot):
        col = np.expand_dims(np.array(colors[i]), axis=0)
        xi = [Xt2[j,0] for j in range(len(Xt2[:,0])) if df['VOC'].iloc[j] == u]
        yi = [Xt2[j,1] for j in range(len(Xt2[:,1])) if df['VOC'].iloc[j] == u]
        ax2.scatter(xi, yi, c=col, s=60, edgecolors='k',label=str(u))
    
    ax2.set_xlabel("PC1")
    ax1.set_ylabel("PC2")
    ax2.set_title('Main features')
 
plt.legend()
plt.show()
```

## Prediction

–°–ø–µ—Ä–≤–∞ —Å–º–æ—Ç—Ä–∏–º –Ω–∞ –ø–æ–ª–Ω–æ–º –Ω–∞–±–æ—Ä–µ.

–ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Ç—É—Ç –Ω–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ –≤—Å–µ —á–∏—Å–ª–æ–≤—ã–µ, –Ω–æ –ø—Ä–∏—à–ª–æ—Å—å –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å  –æ—Ç–≤–µ—Ç—ã, —Ç–∞–∫ –∫–∞–∫ –∏–Ω–∞—á–µ –ø—É—Ç–∞–ª–∏—Å—å —Ü–≤–µ—Ç–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö.

```python
#from sklearn.datasets import make_classification
from sklearn.multiclass import OneVsOneClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.svm import LinearSVC
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

# from https://habr.com/ru/articles/456294/
# https://www.dmitrymakarov.ru/data-analysis/encoding-10/#4-bazovye-metody-kodirovaniya
# https://github.com/scikit-learn/scikit-learn/issues/5930

le = LabelEncoder()
res = le.fit_transform(df['VOC'])

df['Labels'] = pd.Series(res)
#print(df.head())

skpca = sk_pca(n_components=2)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0) 
sc = preprocessing.MaxAbsScaler() 

X_train = skpca.fit_transform(sc.fit_transform(df_train.values[:, 2:-1]))
#X_test = sc.transform(df_test.values[:, 2:])
Y_train = df_train['Labels']
#Y_train = df_train.values[:, 0]
X_test = skpca.fit_transform(sc.transform(df_test.values[:, 2:-1]))
y_test = df_test['Labels']
#y_test = df_test.values[:, 0]
```

### One-vs-one

```python
clf = OneVsOneClassifier(LinearSVC(random_state=0)).fit(X_train, Y_train) # LinearSVC(dual="auto", random_state=0)
y_preds_ovo = clf.predict(X_test)

print(classification_report(y_test, y_preds_ovo))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_preds_ovo)
#print('–û—Ü–µ–Ω–∫–∞ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: \n', cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(xticks_rotation=45)
plt.title('Confusion Matrix Display (Test set)')
plt.show()
```

#### Visualization

```python
# Create color maps
colors = [plt.cm.jet(float(i)/len(voc)) for i in range(len(voc))]

_, ax = plt.subplots()

#  classmethod from_estimator(estimator, X, *, grid_resolution=100, eps=1.0, plot_method='contourf', response_method='auto', class_of_interest=None, xlabel=None, ylabel=None, ax=None, **kwargs)[source]¬∂

DecisionBoundaryDisplay.from_estimator(
    clf,
    X_train, # y_preds_ovo ?
    cmap=ListedColormap(colors), #cmap_light, colors, #
    ax=ax,
    xlabel='PC1',
    ylabel='PC2',
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
)

sns.scatterplot(
    x=X_train[:,0],
    y=X_train[:,1],
#    hue=['Ethanol', 'Ethylene', 'Ammonia', 'Acetaldehyd', 'Acetone', 'Toluene'],
    hue=Y_train, # =df_train['VOC'], #  = df_train.values[:,0], #
#    hue=df_train.values[:,0], # iris.target_names[y],
#    hue=df_train.loc[:,'VOC'], # iris.target_names[y],
    palette=colors, #cmap_bold,
    alpha=1.0,
    edgecolor="black",
#    legend=False,
)
# voc = {1: 'Ethanol', 2: 'Ethylene', 3: 'Ammonia', 4: 'Acetaldehyd', 5: 'Acetone', 6: 'Toluene'}

sns.scatterplot(
    x=X_test[:,0],
    y=X_test[:,1],
    hue=y_test, # =df_test['VOC'], # df_test.values[:,0], # iris.target_names[y],
    palette=colors, #cmap_bold,
    alpha=0.3,
    edgecolor="black",
    legend=False,
)

#plt.legend(labels=le.inverse_transform([i for i in range(len(voc))]))

plt.title("One-vs-one classification")

plt.show()
```

## –ú–µ—Ç–æ–¥ k-–±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π (kNN)





```python
from sklearn.neighbors import KNeighborsClassifier

'''
clf = OneVsOneClassifier(LinearSVC(random_state=0)).fit(X_train, Y_train) # LinearSVC(dual="auto", random_state=0)
y_preds_ovo = clf.predict(X_test)

print(classification_report(y_test, y_preds_ovo))

'''

skpca = sk_pca(n_components=2)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0) 
sc = preprocessing.MaxAbsScaler() 

X_train = skpca.fit_transform(sc.fit_transform(df_train.values[:, 2:-1]))
Y_train = df_train['Labels']
X_test = skpca.fit_transform(sc.transform(df_test.values[:, 2:-1]))
y_test = df_test['Labels']

n_neighbors = 15
knn = KNeighborsClassifier(n_neighbors, weights='distance') # , weights='uniform'
knn.fit(X_train, Y_train)
y_preds_knn = knn.predict(X_test)
print(classification_report(y_test, y_preds_knn))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_preds_knn)
#print('–û—Ü–µ–Ω–∫–∞ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: \n', cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot(xticks_rotation=45)
plt.title('Confusion Matrix Display (Test set)')
plt.show()
```

#### Visualization

```python
# Create color maps
colors = [plt.cm.jet(float(i)/len(voc)) for i in range(len(voc))]

_, ax = plt.subplots()
DecisionBoundaryDisplay.from_estimator(
    knn,
    X_train,
    cmap=ListedColormap(colors), #cmap_light,
    ax=ax,
    xlabel='PC1',
    ylabel='PC2',
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
)

sns.scatterplot(
    x=X_train[:,0],
    y=X_train[:,1],
    hue=Y_train, # iris.target_names[y],
    palette=colors, #cmap_bold,
    alpha=1.0,
    edgecolor="black",
)

sns.scatterplot(
    x=X_test[:,0],
    y=X_test[:,1],
    hue=y_test, # iris.target_names[y],
    palette=colors, #cmap_bold,
    alpha=0.3,
    edgecolor="black",
    legend=False,
)

plt.title(f"{len(voc)}-Class classification (k = {n_neighbors})")
plt.show()
```

–ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è —Ç–∞–∫–∞—è –∫—Ä–∞—Å–∏–≤–µ–µ –≤–∏–∑—É–∞–ª—å–Ω–æ, –Ω–æ —Ç–æ—á–Ω–æ—Å—Ç—å —Ç—É—Ç –Ω–∏–∂–µ –ø–æ –º–µ—Ç—Ä–∏–∫–∞–º.

–•–æ—Ç–µ–ª –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ, —á—Ç–æ –±—É–¥–µ—Ç –µ—Å–ª–∏ —ç—Ç–æ –¥–µ–ª–∞—Ç—å —Ç–æ–ª—å–∫–æ –Ω–∞ –Ω–∞–±–æ—Ä–µ –æ—Å–Ω–æ–≤–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –Ω–æ –∫–∞–∂–µ—Ç—Å—è —Ç–∞–º –±—É–¥–µ—Ç –Ω–µ–≤–∞–∂–Ω–æ.

–ü–æ—ç—Ç–æ–º—É –ø–æ–ø—Ä–æ–±—É–µ–º –ø–æ–¥–æ–±—Ä–∞—Ç—å –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã.


–í –º–æ–¥–µ–ª–∏ kNN –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –æ–±—ã—á–Ω–æ –≤—ã—Å—Ç—É–ø–∞—é—Ç:

–∞) –ß–∏—Å–ª–æ —Å–æ—Å–µ–¥–µ–π (n_neighbors) ‚Äî —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º —Å—Ç–æ–∏—Ç —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –∏ –Ω–µ –±–æ—è—Ç—å—Å—è —Å—Ç–∞–≤–∏—Ç—å –±–æ–ª—å—à–∏–º;

–±) –°—Ç–µ–ø–µ–Ω—å –¥–ª—è –º–µ—Ç—Ä–∏–∫–∏ –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ (p) ‚Äî 1 –∏–ª–∏ 2 –∫—Ä–∞–π–Ω–µ —á–∞—Å—Ç–æ –±—É–¥—É—Ç –ª—É—á—à–∏–º–∏.

We will use three hyperparamters- n-neighbors, weights and metric.

*    n_neighbors: Decide the best k based on the values we have computed earlier.
*    weights: Check whether adding weights to the data points is beneficial to the model or not. 'uniform' assigns no weight, while 'distance' weighs points by the inverse of their distances meaning nearer points will have more weight than the farther points.
*    metric: The distance metric to be used will calculating the similarity.

–ò—Å—Ç–æ—á–Ω–∏–∫: https://www.kaggle.com/code/arunimsamudra/k-nn-with-hyperparameter-tuning

(–∏ —Ç—É—Ç –µ—Å—Ç—å –µ—â–µ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è cross_val_score)


–°–ø–µ—Ä–≤–∞ –ø–æ–ø—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ —Å–æ—Å–µ–¥–µ–π - —ç—Ç–æ –æ—Å–Ω–æ–≤–Ω–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä –º–µ—Ç–æ–¥–∞.

```python
from sklearn import metrics

mean_acc = np.zeros(20)
for i in range(1,21):
    #Train Model and Predict  
    knn = KNeighborsClassifier(n_neighbors = i).fit(X_train, Y_train)
    yhat= knn.predict(X_test)
    mean_acc[i-1] = metrics.accuracy_score(y_test, yhat)

mean_acc

loc = np.arange(1,21,step=1.0)
plt.figure(figsize = (10, 6))
plt.plot(range(1,21), mean_acc)
plt.xticks(loc)
plt.xlabel('Number of Neighbors ')
plt.ylabel('Accuracy')
plt.show()
```

–¢–æ—á–Ω–æ—Å—Ç—å –ø–æ–≤—ã—à–µ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –º–µ–Ω–µ–µ 10.

–¢–µ–ø–µ—Ä—å –ø–æ–¥–±–µ—Ä–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã.

```python
from sklearn.model_selection import GridSearchCV

grid_params = { 'n_neighbors' : [i*2 for i in range(1, 11, 2)],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan'],
              'algorithm' : ['ball_tree', 'kd_tree', 'brute']}

gs = GridSearchCV(KNeighborsClassifier(), grid_params, verbose = 1, cv=3, n_jobs = -1)

# fit the model on our train set
g_res = gs.fit(X_train, Y_train)

# find the best score
g_res.best_score_
# get the hyperparameters with the best score
g_res.best_params_
```

```python
# use the best hyperparameters
n_neighbors = 2

knn = KNeighborsClassifier(n_neighbors = n_neighbors, weights = 'distance', algorithm = 'ball_tree',metric = 'minkowski')
knn.fit(X_train, Y_train)
```

```python
y_preds_knn = knn.predict(X_test)
print(classification_report(y_test, y_preds_knn))
```
